```python
import pandas as pd

dados = pd.read_csv('titanic.csv') #importando o arquivo csv

def mulheres_sobreviventes_primeira_classe(dados): #definindo a função
    filter1 = dados[['Name', 'Age','Sex','Pclass','Survived']] #filtrando as colunas que serão utilizadas
    filter1 = filter1.loc[(filter1['Sex'] == 'female')  & (filter1['Pclass'] == 1) & (filter1['Survived'] == 1)] #filtrando os dados
    filter1 = filter1.sort_values(['Name'], ascending=True) #ordenando os dados por ordem alfabética
    filter1.to_csv('mulheres_sobreviventes_primeira_classe.csv', index=False) #salvando o arquivo csv
    print(filter1) #exibindo o resultado

def criancas_sobreviventes(dados):
    filter_children = dados[(dados['Age'] < 12) & (dados['Survived'] == 1)]  
    filter_children = filter_children[['Name', 'Age', 'Survived']]  # Selecionando as colunas relevantes
    filter_children = filter_children.sort_values(['Age'], ascending=True)  # Ordenando por idade
    filter_children.to_csv('criancas_sobreviventes.csv', index=False) 
    print(filter_children) 

def idosos_a_bordo(dados):
    filter_idosos = dados[dados['Age'] > 60]  
    filter_idosos = filter_idosos[['Name', 'Age', 'Survived']] 
    filter_idosos = filter_idosos.sort_values(['Name'], ascending=True)
    filter_idosos.to_csv('idosos_a_bordo.csv', index=False)
    print(filter_idosos)

# Chamando as funções para exibir os resultados
mulheres_sobreviventes_primeira_classe(dados)
criancas_sobreviventes(dados)
idosos_a_bordo(dados)
```
